```python
import pandas as pd
pd.set_option('display.max.colwidth', 100)
resultprac = pd.read_csv("./crawlingresult.csv", encoding = 'ms949')

deldata = resultprac['평점'].isin(['평없음'])
realdeldata = resultprac [~deldata]

realdeldata.dtypes

realdeldata["평점"] = realdeldata["평점"].str.replace(",", "").astype(int)
realdeldata["리뷰갯수"] = realdeldata["리뷰갯수"].str.replace(",", "").astype(int)
realdeldata["가격"] = realdeldata["가격"].str.replace(",", "").astype(int)

realdeldata

answer = realdeldata.loc[(realdeldata.평점 > 90 ) & 
                         (realdeldata.리뷰갯수 > 400) &
                        (realdeldata.스펙갯수>=5)&
                        (realdeldata.가격>150000)&
                        (realdeldata.가격<300000)]


answer = answer.sort_values("가격")
answer.reset_index(drop = True)
```
